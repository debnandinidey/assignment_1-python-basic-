1. What does RGBA stand for?

answer:RGBA stands for Red, Green, Blue, and Alpha. It is a color model that represents colors using a combination of red, green, and blue components, along with an additional alpha component for transparency control. 

2. From the Pillow module, how do you get the RGBA value of any images?

answer: we can use the getpixel() method

```python
from PIL import Image

# Open the image
image = Image.open("kevin-crosby-NBwwEsTFgJU-unsplash.jpg")

# Get the width and height of the image
width, height = image.size

# Example: Get the RGBA value of the pixel at the center of the image
x = width // 2
y = height // 2
rgba = image.getpixel((x, y))

# Print the RGBA value
print(rgba)
```

4. Use your image and load in notebook then, How can you find out the width and height of an
Image object?

anwer:we can use the size attribute of the Image object.

```python
from PIL import Image
import matplotlib.pyplot as plt

# Open the image
image = Image.open("kevin-crosby-NBwwEsTFgJU-unsplash.jpg")

# Get the width and height
width, height = image.size

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()

# Print the width and height
print("Width:", width)
print("Height:", height)
```

5. What method would you call to get Image object for a 100100 image, excluding the lower-left
quarter of it?

answer: we can use crop() method of the image object.

```python
from PIL import Image

# Open the image
image = Image.open("kevin-crosby-NBwwEsTFgJU-unsplash.jpg")

# Crop the image
cropped_image = image.crop((0, 50, 50, 100))

# Display the cropped image
cropped_image.show()
```
